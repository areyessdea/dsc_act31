```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn import metrics
from yellowbrick.cluster import KElbowVisualizer, InterclusterDistance
from yellowbrick.features import RadViz
```

## Parte 1

```python
articles_df = pd.read_csv("data/articles.csv", encoding="utf-8")
```

```python
# Se vectoriza y eliminan stop words y términos que aparecen en menos de 5 artículos
# y en más del 50% de los artículos
vectorizer = TfidfVectorizer(max_df=0.5, min_df=5,stop_words="english")
X = vectorizer.fit_transform(articles_df['content'].values.astype('U'))
print(f"Artículos: {X.shape[0]}, Términos: {X.shape[1]}")
```

```python
print(f"Valores que no son cero: {X.nnz / np.prod(X.shape):.3f}")
```

```python
visualizer = RadViz(size=(600, 400))
model =KMeans(max_iter=100, n_init=5, random_state=123)
visualizer = KElbowVisualizer(model, k=(6,22), metric='silhouette')
visualizer.fit(X)
visualizer.show();
```

## Parte 2

### Centroides de los clusters (paso 1)

```python
true_k = 18
model = KMeans(n_clusters=true_k, max_iter=100, n_init=5, random_state=123).fit(X)
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
order_centroids
```

### Principales palabras por cluster (paso 2 y 3)

```python
print("Principales palabras por cluster")
terms = vectorizer.get_feature_names_out()
for i in range(true_k):
    print("Cluster %d:" % i, end=''),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind], end=''),
    print()
```

### Limitando la cantidad de palabras (4)

```python
# Se vectoriza y eliminan stop words y términos que aparecen en menos de 5 artículos
vectorizer = TfidfVectorizer(max_features= 1000,max_df=0.5, min_df=5,stop_words="english")
X = vectorizer.fit_transform(articles_df['content'].values.astype('U'))
print(f"Artículos: {X.shape[0]}, Términos: {X.shape[1]}")
```

```python
print(f"Valores que no son cero: {X.nnz / np.prod(X.shape):.3f}")
```

```python
true_k = 18
model = KMeans(n_clusters=true_k, max_iter=100, n_init=5, random_state=123).fit(X)
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
print("Principales palabras por cluster")
terms = vectorizer.get_feature_names_out()
for i in range(true_k):
    print("Cluster %d:" % i, end=''),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind], end=''),
    print()
```

Se observan cambios importantes en temas y palabras

### Muestra aleatoria de artículos por cluster (paso 5)

```python
df = articles_df.copy()
clust_labels = model.predict(X)
cent = model.cluster_centers_
kmeans_labels = pd.DataFrame(clust_labels)
df.insert((df.shape[1]),'clusters',kmeans_labels)
```

```python
new_df = df.groupby('clusters').sample(n=5)
new_df.loc[:,['headline', 'section_name', 'clusters']]
```
